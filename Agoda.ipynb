```python
time.sleep(3)  # Allow 2 seconds for the web page to (open depends on you)
scroll_pause_time = 10  # You can set your own pause time. dont slow too slow that might not able to load more data
screen_height = driver.execute_script("return window.screen.height;")  # get the screen height of the web
i = 1
page=1
data=[]

while True:
    # scroll one screen height each time
    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))
    i += 1
    time.sleep(scroll_pause_time)
    # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
    scroll_height = driver.execute_script("return document.body.scrollHeight;")
    print(scroll_height)
    # Break the loop when the height we need to scroll to is larger than the total scroll height
    if (screen_height) * i > scroll_height:
        soup=BeautifulSoup(driver.page_source,'html.parser')
        elements=soup.find_all('li',class_='PropertyCard PropertyCardItem')
        for x in elements:
            data.append(x)
        print(len(data))
        next_page_button = WebDriverWait(driver, 120).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[11]/div/div[2]/div[1]/div/div[2]/div[2]/div/div/button/div/div/span[text()='Next']")))
        next_page_button.click()
        print(page)
        page=page+1
        time.sleep(scroll_pause_time)
        i=1
        if page==35:
            break 
```

```python
csv_nm=r"C:\Users\chemi\Desktop\save_csv\Agoda_hotels_sample.csv"
fields=["Hotle_name",'Address','Price']
with open(csv_nm, 'w', encoding="utf-16", newline='') as file:
    writer = csv.DictWriter(file, fieldnames=fields, delimiter =";")
    writer.writeheader()
    agoda = {}
    for i in data:
        try:
            hotel_name=i.find_all('h3',class_="sc-kEjbxe sc-iqHYGH bpoMrU bCVHpR")
            agoda["Hotle_name"]=hotel_name[0].text
        except:
            agoda["Hotle_name"]=np.nan
            
        try:
            address=i.find_all('span',class_="sc-kEjbxe sc-iqHYGH bpoMrU hGjjJo sc-dQppl cQZIoF")
            agoda["Address"]=address[0].text.split('-')[0]
        except:
            agoda["Address"]=np.nan
            
        try:
            price=i.find_all('span',class_="PropertyCardPrice__Value")
            agoda["Price"]=price[0].text
        except:
            agoda["Price"]=np.nan
            
            
            
        writer.writerow(agoda)           
print(agoda)
        
file.close() 
    
```
